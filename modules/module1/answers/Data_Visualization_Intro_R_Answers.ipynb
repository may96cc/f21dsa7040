**Activity 2:** Fill the missing NA values in the dataset.

```r
## Your answer for activity 2 goes here ---

# You can do zero, mean, median, etc. 
# This one na.locf to replace each NA with the most recent non-NA. 

# Since different food groups may have very different protein, fat, 
# vitamins, etc. content, mean or median may not work well. 

# Food items are grouped in the data set, so picking the same value 
# for food items that are close to each other is *probably* ok. 

library(zoo)
USDA = na.locf(USDA)
USDA <- lapply(USDA, function(x) as.numeric(x))
USDA = as.data.frame(USDA)
```

**Alternative codes to print the actual pair names:**

```r
all_pairs <- combn(3:16, 2)
for (i in 1:ncol(all_pairs)) {
    idx1 <- all_pairs[1,i]
    idx2 <- all_pairs[2,i]
    corr <- cor(USDA[,idx1], USDA[,idx2])
    if (corr > 0.5) cat(colnames(USDA)[idx1], colnames(USDA)[idx2], fill=TRUE)
}
```

**Activity 4:** Fit a linear regression model between Calories and TotalFat.

```r
## Your answer for activity 4 goes here ---
lm(Calories ~ TotalFat, data = USDA)
lm(TotalFat ~ Calories, data = USDA)
```

**Activity 5:** Plot a grid of **scatter plots** for the 4 pairs of correlations identified in activity 3.
Make sure to add a **regression line**.

```r
## Your answer for activity 5 goes here ---

# Create a scatterplot with Calories and TotalFat as inputs for x and y axis
p1 <- ggplot(USDA, aes(Calories, TotalFat,color=SaturatedFat)) + geom_point() +
    # add a linear regression line and set minimum x,y limits on the plot
    geom_smooth(method = lm, se = FALSE, color="orange")

# Create it with Calories, SaturatedFat
p2 <- ggplot(USDA, aes(Calories, SaturatedFat,color=Cholesterol)) + geom_point() +
    geom_smooth(method = lm, se = FALSE, color="orange")


# Create it with TotalFat, SaturatedFat
p3 <- ggplot(USDA, aes(TotalFat, SaturatedFat,color=Cholesterol)) + geom_point() +
    geom_smooth(method = lm, se = FALSE, color="orange")

# Create it with Carbohydrate, Sugar
p4 <- ggplot(USDA, aes(Carbohydrate, Sugar,color=Calories)) + geom_point() +
    geom_smooth(method = lm, se = FALSE, color="orange")

grid.arrange(p1, p2, p3, p4)
```

**Activity 6:** Plot a scatter plot between Calories and TotalFat. Plot the points in green color and transparent. Adjust the breaks for x and y axes. 

```r
## Your answer for activity 6 goes here ---

ggplot(USDA,aes(TotalFat, Calories)) + geom_point(color="green",alpha = 0.1) +
    # add black and white theme
    theme_bw() +
    # adjust the limits on the axes
    scale_x_continuous(breaks = seq(-10, 110, 10)) + scale_y_continuous(breaks = seq(0, 1000, 100)) +
    # more axis stuff
    expand_limits(x = c(-10,110), y = c(0,1000)) +
    # add title 
    labs(title = "USDA") + geom_smooth(method = lm, se = FALSE,color = "orange")
```
